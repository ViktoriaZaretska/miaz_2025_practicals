```python
import pandas as pd
```

```python
df = pd.read_csv('World_Army_Dataset.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['activeDuty'], bins=10, kde=True)
plt.show()
```

```python
sns.boxplot(x=df['activeDuty'])
plt.show()
```

```python
sns.scatterplot(x=df['activeDuty'], y=df['total'])
plt.show()
```

```python
correlation_matrix = df.select_dtypes(include='number').corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```
